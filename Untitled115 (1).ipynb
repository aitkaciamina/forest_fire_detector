# Import necessary libraries:

```python
import matplotlib.pyplot as plt
import seaborn as sns

import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D , MaxPool2D , Flatten , Dropout 
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam 
from sklearn.metrics import classification_report,confusion_matrix

import tensorflow as tf

import cv2
import os

import numpy as np
```

```python
labels=['with_fire','without_fire']
img_size=256
def get_data(data_dir):
    data = [] 
    for label in labels: 
        path = os.path.join(data_dir, label)
        class_num = labels.index(label)
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path, img))[...,::-1] #convert BGR to RGB format
                resized_arr = cv2.resize(img_arr, (img_size, img_size)) # Reshaping images to preferred size
                data.append([resized_arr, class_num])
            except Exception as e:
                print(e)
    return np.array(data)
```

# Load the data:

```python
#load the data from the folder Fire
data=tf.keras.utils.image_dataset_from_directory('C:/Users/user/Pictures/Fire')
```

```python
print(data)
```

```python
#convert it into numpy_iterator to access the data pipeline
data_iterator=data.as_numpy_iterator()
#grab one batch back
batch=data_iterator.next()
```

## viewing 4 random images:

```python
#see the indices of the classes
fig,ax=plt.subplots(ncols=4,figsize=(25,25))
for idx ,img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch[1][idx])
#1:fire
#0:non fire
```

```python
#Scale the data
data=data.map(lambda x,y:(x/255 , y))
```

```python
data.as_numpy_iterator().next()
```

# Splitting the data:

```python
#returns the batches
train_size=int(len(data)*.7)
val_size=int(len(data)*.2)
test_size=int(len(data)*.1)+1
```

```python
train=data.take(train_size)
val=data.skip(train_size).take(val_size)
test=data.skip(train_size+val_size).take(test_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D ,MaxPooling2D , Dense,Flatten,Dropout
model=Sequential()
from tensorflow.keras.models import Model
from tensorflow.keras.backend import sigmoid
```

# Train the model:

```python
model=Sequential()
```

```python
model.add(Conv2D(16,(3,3),1,activation='relu',input_shape=(256,256,3)))
model.add(MaxPool2D())

# Convolutional layer and maxpool layer 2
model.add(Conv2D(32,(3,3),1,activation='relu'))
model.add(MaxPool2D())

# Convolutional layer and maxpool layer 3
model.add(Conv2D(64,(3,3),1,activation='relu'))
model.add(MaxPool2D(2,2))

# This layer flattens the resulting image array to 1D array
model.add(Flatten())

# Hidden layer with 512 neurons and Rectified Linear Unit activation function 
model.add(Dense(256,activation='relu'))

# Output layer with single neuron which gives 0 for Cat or 1 for Dog 
#Here we use sigmoid activation function which makes our model output to lie between 0 and 1
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
m=model.fit(train, validation_data = val, epochs = 5)
```

```python
m.history
```

```python
model.summary()
```

# Evaluate the model:

```python
fig=plt.figure()
plt.plot(m.history['loss'],color='teal',label='loss')
plt.plot(m.history['val_loss'],color='orange',label='val_loss')
fig.suptitle('loss',fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig=plt.figure()
plt.plot(m.history['accuracy'],color='teal',label='loss')
plt.plot(m.history['val_accuracy'],color='orange',label='val_accuracy')
fig.suptitle('accuracy',fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
from tensorflow.keras.metrics import Precision , Recall,BinaryAccuracy
```

```python
pre=Precision()
re=Recall()
ba=BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    X,y=batch
    yhat=model.predict(X)
    pre.update_state(y,yhat)
    re.update_state(y,yhat)
    ba.update_state(y,yhat)
```

```python
print(f'Precision:{pre.result().numpy()},Recall:{re.result().numpy()},Accuracy:{ba.result().numpy()}')
```
